```python
import pandas as pd
import numpy as np

def generate_synthetic_dataset(num_samples) -> pd.DataFrame:

    time_of_day_choices = ["Morning", "Noon", "Night"]
    activity_choices = ["Swimming", "Resting", "Running", "Walking", "Driving"]
    weather_choices = ["Clear", "Rainy", "Cold"]
    day_of_week_choices = ["Weekday", "Holiday"]
    recent_events_choices = ["YES", "NO"]

    np.random.seed(42)

    time_of_day = np.random.choice(time_of_day_choices, size=num_samples)
    activity = np.random.choice(activity_choices, size=num_samples)
    weather = np.random.choice(weather_choices, size=num_samples)
    day_of_week = np.random.choice(day_of_week_choices, size=num_samples)
    recent_events = np.random.choice(recent_events_choices, size=num_samples)

    age = np.random.randint(10, 101, size=num_samples)
    heart_rate = np.random.randint(30, 201, size=num_samples)
    battery_status = np.random.randint(1, 101, size=num_samples)
    sleep_score = np.random.randint(1, 101, size=num_samples)

    df = pd.DataFrame({
        "Time of Day": time_of_day,
        "Activity": activity,
        "Age": age,
        "Heart Rate": heart_rate,
        "Weather": weather,
        "Day of Week": day_of_week,
        "Recent Events": recent_events,
        "Battery Status": battery_status,
        "Sleep Score": sleep_score
    })

    df["Daily Activity"] = np.where(df["Activity"] != "Resting", 1, 0)
    df["Sleep"] = np.where(df["Sleep Score"] < 60, 1, 0)
    df["Running"] = np.where(df["Activity"] == "Running", 1, 0)
    df["Swimming"] = np.where(df["Activity"] == "Swimming", 1, 0)
    df["Calories Burn"] = np.where(df["Activity"].isin(["Running", "Walking", "Swimming"]), 1, 0)
    df["Heart Rate (show)"] = np.where((df["Heart Rate"] < 60) | (df["Heart Rate"] > 100), 1, 0)
    df["Steps"] = np.where(df["Activity"].isin(["Running", "Walking"]), 1, 0)
    df["Consumed"] = np.where(
        (df["Activity"] == "Resting") & (df["Time of Day"].isin(["Noon", "Night"])), 1, 0
    )
    weather_output_values = np.where(df["Weather"] != "Clear", 1, 0)
    df["Weather (show)"] = weather_output_values
    df["Chance of Rain"] = np.where(df["Weather"] == "Rainy", 1, 0)
    df["Events"] = np.where(df["Recent Events"] == "YES", 1, 0)
    df["Battery"] = np.where(df["Battery Status"] < 20, 1, 0)

    return df
```

```python
NUM_SAMPLES = 1000
df = generate_synthetic_dataset(NUM_SAMPLES)

print("First 5 rows of the synthetic dataset:")
print(df.head())

output_filename = "../data/synthetic/synthetic_multilabel_dataset.csv"
df.to_csv(output_filename, index=False)
print(f"\nSaved {NUM_SAMPLES} samples to '{output_filename}'.")
```
